```python
# Irisのエータフレームのヒストグラムを作成
Iris.hist(bins=10, figsize=(20,15), color = 'teal')
```

```python
# ヒストグラムを作成
fig = plt.figure(figsize=(25,10))
p1 = fig.add_subplot(2,2,1)
p1.hist(Iris.petal_length[Iris.species == 'setosa'], bins=10, alpha = .4)
p1.hist(Iris.petal_length[Iris.species == 'versicolor'], bins=10, alpha = .4)
p1.hist(Iris.petal_length[Iris.species == 'virginica'], bins=10, alpha = .4)
plt.title('petal length')
plt.xlabel('Cm Measurement')
plt.ylabel('Count')
labels = ["Iris-setosa", "Iris-versicolor", "Iris-virginica"]
plt.legend(labels)

p2 = fig.add_subplot(2,2,2)
p2.hist(Iris.petal_width[Iris.species == 'setosa'], bins=10, alpha = .4)
p2.hist(Iris.petal_width[Iris.species == 'versicolor'], bins=10, alpha = .4)
p2.hist(Iris.petal_width[Iris.species == 'virginica'], bins=10, alpha = .4)
plt.title('petal Width Cm')
plt.xlabel('Cm Measurement')
plt.ylabel('Count')
labels = ["Iris-setosa", "Iris-versicolor", "Iris-virginica"]
plt.legend(labels)

p3 = fig.add_subplot(2,2,3)
p3.hist(Iris.sepal_length[Iris.species == 'setosa'], bins=10, alpha = .4)
p3.hist(Iris.sepal_length[Iris.species == 'versicolor'], bins=10, alpha = .4)
p3.hist(Iris.sepal_length[Iris.species == 'virginica'], bins=10, alpha = .4)
plt.title('Sepal Length Cm')
plt.xlabel('Cm Measurement')
plt.ylabel('Count')
labels = ["Iris-setosa", "Iris-versicolor", "Iris-virginica"]
plt.legend(labels)
plt.title('Sepal Length Cm')

p4 = fig.add_subplot(2,2,4)
p4.hist(Iris.petal_width[Iris.species == 'setosa'], bins=10, alpha = .4)
p4.hist(Iris.petal_width[Iris.species == 'versicolor'], bins=10, alpha = .4)
p4.hist(Iris.petal_width[Iris.species == 'virginica'], bins=10, alpha = .4)
plt.title('Sepal Width Cm')
plt.xlabel('Cm Measurement')
plt.ylabel('Count')
labels = ["Iris-setosa", "Iris-versicolor", "Iris-virginica"]

plt.legend(labels)
plt.subplots_adjust(wspace=.1, hspace=.3)
plt.show()
```
